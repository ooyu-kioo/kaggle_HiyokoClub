## データの整形

#### データ型変更・列名変更・並び替えなど

```python
# datetime列をindexにする
df.set_index('datetime', inplace=True)
df.head()
```

```python
# indexの確認
df.index
```

```python
# カラム名を変更する
df.rename(columns={'y': 'sales'}, inplace=True)
df.head()
```

```python
# 'sales'列を降順で並び替える
df.sort_values(by='sales', ascending=False).head() # ascending=Trueで昇順
```

```python
# sort_valuesは複数の列に対しても実行できる
# 以下のコードでは、'sales'が同じ場合に'temperature'が高いものから順に並ぶ
df.sort_values(['sales', 'temperature'], ascending=False).head(15)
# ascending=Trueで昇順
```

```python
# indexのデータ型を確認
df.index # dtype='object'であることがわかる
```

```python
# index(datetime)のdtype='object'を、dtype='datetime64[ns]'に変更
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
```

```python
df.index # dtype='datetime64[ns]'になった！
```

#### indexをdatetime型に変換したことで、できること

```python
# indexに対してsortを行う
df.sort_index().head() # object型のままだと正しくsortされない
```

```python
# resampleメソッドで、日単位や月単位で簡単に集計できる
df.resample('M').mean() # 月単位で平均値を出力
```

```python
# 簡単にmonth列やday列をつくれるようになる
df['month'] = list(pd.Series(df.index).apply(lambda x: x.month))
df['day'] = list(pd.Series(df.index).apply(lambda x: x.day))

df.head()
```

```python
# cutメソッドでヒストグラムのピン指定的な処理ができる
labels = ['上旬', '中旬', '下旬']
df['period'] = pd.cut(list(df['day']), bins=[0,10,20,31], labels=labels, right=True)
# 0<day<=10, 10<day<=20, 20<day<=31

df.head()
```

#### データの欠損状態の確認

* データに欠損値NaN(Not a Number)があると色々と問題に…
* →解決しよう！

```python
# 列単位で、欠損値NaNが入っている個数をカウントする
# (正確には、isnull()でtrueが帰ってくる数をカウントしている)
df.isnull().sum()
```

```python
# 1つでもNaNが含まれる行だけを抽出
print(df[df.isnull().any(axis=1)].shape)
df[df.isnull().any(axis=1)].head()
```

```python
# 'payday'列にあるNaNを'0.0'に置き換える
df.fillna(value={'payday': 0.0}, inplace=True)
df.head()
```

```python
df.isnull().sum() # 'payday'が0になった！
```

```python
# 'kcal'列にNaNがある行を削除する
df.dropna(subset=['kcal'], axis=0, inplace=True)
print(df.shape) # 207-166=41行のデータを削除した
```

```python
df.isnull().sum() # 'kcal'が0になった！
```

```python
# 'precipitation'列の'--'を0に置き換える
df['precipitation'] = df['precipitation'].replace('--', 0).astype(float)
df.head()
```

```python
# maskメソッドを使う例。
# 'sales'列が80よりも大きければ、その値を100に置換する。
pd.DataFrame(df['sales'].mask(df['sales'] > 80, 100)).head()
    # df.head()を見た感じ置き換わってないけど…
```

```python
# 'remarks'はデータとして不要な気がするので、データを列ごと削除
df.drop(['remarks'], axis=1, inplace=True)
df.head()
```

## 集計

```python
# weather列の集計
df['weather'].value_counts()
```

```python
# groupbyメソッドで、'week'列ごとに'soldout'の数をカウントする
df.groupby(['week'])['soldout'].count()
```

```python
# groupbyメソッドは、複数列に対しても行える
# groupbyメソッドで、'month', 'period'列ごとに'sales'の数を合計する
df.groupby(['month', 'period'])['sales'].sum()
```

```python
# weatherごとにtemperatureの平均値を出す
df.groupby(['weather'])['temperature'].mean()
```

```python
# 全行との差分が欲しい時は.diff()を使う
df['temperature_diff'] = df['temperature'].diff(periods=1)
df[['temperature', 'temperature_diff']].head()
```

```python
# 'temperature'列について、頭からwindowサイズ3で移動平均を計算
df['temperature_rolling_mean'] = df['temperature'].rolling(window=3).mean()
df[['temperature', 'temperature_diff', 'temperature_rolling_mean']].head()
```

```python
# pct_changeで、変化率を計算
df['temperature_pct_change'] = df['temperature'].pct_change()
df[['temperature', 'temperature_diff', 'temperature_rolling_mean', 'temperature_pct_change']].head()
```

```python
# 'temperature'列に1つでもNaNがある行を削除
df.dropna(subset=['temperature_diff', 'temperature_rolling_mean', 'temperature_pct_change'], axis=0, inplace=True)
```

#### 前処理をしたデータはこの後どう活用するのか？

* 例えば、'sales'を「目的変数」、それ以外の「説明変数」として、目的変数を予想する"Model"を作成するなど
    * より目的変数に"効く"説明変数をつくることを「特徴量エンジニアリング」という
    
* Pandasを使った前処理とは、NaNや無関係な説明変数を削除しつつ、"効く"特徴量を探す(作成する)作業のこと
    * 地味だが、Modelの精度を上げるうえでは非常に大切！
    * データ分析のほとんどの時間はここに使われる
